```python
filenames_0 = [
    "moa.classifiers.lazy.kNN.txt",
    "moa.classifiers.meta.OzaBoost -s 5.txt",
    "moa.classifiers.functions.SGD.txt",
    "moa.classifiers.trees.HoeffdingTree.txt",
    "moa.classifiers.trees.HoeffdingAdaptiveTree.txt"
]

filenames_1 = [
    "moa.classifiers.meta.AdaptiveRandomForest -s 20.txt",
    "moa.classifiers.meta.WeightedMajorityAlgorithm -l functions.SGD,meta.AdaptiveRandomForest.txt",
    "moa.classifiers.meta.WeightedMajorityAlgorithm -l functions.SGD,trees.HoeffdingAdaptiveTree.txt",
    "moa.classifiers.meta.WeightedMajorityAlgorithm -l trees.HoeffdingAdaptiveTree,meta.AdaptiveRandomForest.txt",
    "moa.classifiers.meta.OzaBoost -l trees.HoeffdingAdaptiveTree -s 5.txt",
    "moa.classifiers.meta.AdaptiveRandomForest.txt"
]

def load_file(problem, alg, part):
    path = ""
    if part == 0:
        path = "../experiments/Results_0/Results/kddcup_"
    else: 
        path = "../experiments/Results_1/Results/kddcup_"
    path += problem
    
    path += "/" + alg + ".txt"
    
    return pd.read_csv(path)

knn_http = load_file("http", "moa.classifiers.lazy.kNN", 0)
boostHT_http = load_file("http", "moa.classifiers.meta.OzaBoost -s 5", 0)
svm_http = load_file("http", "moa.classifiers.functions.SGD", 0)
ht_http = load_file("http", "moa.classifiers.trees.HoeffdingTree", 0)
hat_http = load_file("http", "moa.classifiers.trees.HoeffdingAdaptiveTree", 0)

arf20_http = load_file("http", "moa.classifiers.meta.AdaptiveRandomForest -s 20", 1)
svm_arf_http = load_file("http", "moa.classifiers.meta.WeightedMajorityAlgorithm -l functions.SGD,meta.AdaptiveRandomForest", 1)
svm_hat_http = load_file("http", "moa.classifiers.meta.WeightedMajorityAlgorithm -l functions.SGD,trees.HoeffdingAdaptiveTree", 1)
hat_arf_http = load_file("http", "moa.classifiers.meta.WeightedMajorityAlgorithm -l trees.HoeffdingAdaptiveTree,meta.AdaptiveRandomForest", 1)
boostHAT_http = load_file("http", "moa.classifiers.meta.OzaBoost -l trees.HoeffdingAdaptiveTree -s 5", 1)
arf_http = load_file("http", "moa.classifiers.meta.AdaptiveRandomForest", 1)

knn_smtp = load_file("smtp", "moa.classifiers.lazy.kNN", 0)
boostHT_smtp = load_file("smtp", "moa.classifiers.meta.OzaBoost -s 5", 0)
svm_smtp = load_file("smtp", "moa.classifiers.functions.SGD", 0)
ht_smtp = load_file("smtp", "moa.classifiers.trees.HoeffdingTree", 0)
hat_smtp = load_file("smtp", "moa.classifiers.trees.HoeffdingAdaptiveTree", 0)

arf20_smtp = load_file("smtp", "moa.classifiers.meta.AdaptiveRandomForest -s 20", 1)
svm_arf_smtp = load_file("smtp", "moa.classifiers.meta.WeightedMajorityAlgorithm -l functions.SGD,meta.AdaptiveRandomForest", 1)
svm_hat_smtp = load_file("smtp", "moa.classifiers.meta.WeightedMajorityAlgorithm -l functions.SGD,trees.HoeffdingAdaptiveTree", 1)
hat_arf_smtp = load_file("smtp", "moa.classifiers.meta.WeightedMajorityAlgorithm -l trees.HoeffdingAdaptiveTree,meta.AdaptiveRandomForest", 1)
boostHAT_smtp = load_file("smtp", "moa.classifiers.meta.OzaBoost -l trees.HoeffdingAdaptiveTree -s 5", 1)
arf_smtp = load_file("smtp", "moa.classifiers.meta.AdaptiveRandomForest", 1)
```

```python
%matplotlib widget
def plot(dfs, names, title):
    fig, ax = plt.subplots(figsize=(12, 6))
    
    for index, df in enumerate(dfs):
        x = df["learning evaluation instances"]
        y = df["AUC"]
        ax.plot(x, y, label=names[index])
        
    ax.legend()
    ax.set_title(title)
    ax.set_ylabel("AUC")
    ax.set_xlabel("# Instances Seen")
    ax.grid()
    
    return fig
```

```python
plot([svm_arf_http, svm_hat_http, hat_arf_http], ["SVM + ARF", "SVM + HAT", "HAT + ARF"], "Ensemble Classifiers (HTTP)")
```
